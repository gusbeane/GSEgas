```python
from PIL import Image
from IPython.display import display, HTML
import os
import glob

set1_path = './'
set2_path = './'
set3_path = './'
set4_path = './'

# Get the list of images from both sets
set1_images = sorted([img for img in glob.glob('PSB_sfh/*sfh.png')])
set2_images = sorted([img for img in glob.glob('PSB_sfh/*plane.png')])
set3_images = sorted([img for img in glob.glob('PSB_sfh_s0.1/*sfh.png')])
set4_images = sorted([img for img in glob.glob('PSB_sfh_s0.1/*plane.png')])

# Ensure both sets have the same number of images
assert len(set1_images) == len(set2_images), "The two sets must have the same number of images"
assert len(set1_images) == len(set3_images)
assert len(set1_images) == len(set4_images)

# Display the images side by side
html = "<table>"
for img1, img2, img3, img4 in zip(set1_images, set2_images, set3_images, set4_images):
    img1_path = os.path.join(set1_path, img1)
    img2_path = os.path.join(set2_path, img2)
    img3_path = os.path.join(set3_path, img3)
    img4_path = os.path.join(set3_path, img4)
    html += f"<tr><td><img src='{img1_path}'></td><td><img src='{img2_path}'></td><td><img src='{img3_path}'></td><td><img src='{img4_path}'></td></tr>"
html += "</table>"

display(HTML(html))
```

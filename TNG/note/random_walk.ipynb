```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import lognorm, norm, binom
```

```python
p = 0.005
q = 1 - p
```

```python
arr = np.random.rand(10000, 10000)
# print(arr[:10])
advance = arr < p
not_advance = arr >= p

arr[advance] = 1
arr[not_advance] = 0.0000001

# arr[arr < p] = 1
# arr[arr >= p] = 0
val = np.sum(arr, axis=1)

mu, std = np.mean(val), np.std(val)
print(mu, std)
```

```python
bins = np.linspace(mu - 3 *std, mu + 3 * std, 16)
plt.hist(val, bins=bins, density=True);

s, loc, scale = lognorm.fit(val)

print(s, loc, scale)
xlist = np.linspace(mu - 3 *std, mu + 3 * std, 1000)
ylist = lognorm.pdf(xlist, s, loc, scale)
plt.plot(xlist, ylist)
```

```python
X = norm.rvs(size=1000)
print(norm.fit(X))

Y = np.exp(X)
print(lognorm.fit(Y))
print(lognorm.fit(X))
```
